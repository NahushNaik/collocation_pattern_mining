```python
import numpy as np
import pandas as pd
from xml.dom import minidom
```

```python
   
xmldoc = minidom.parse('Zone_3.kml')
kml = xmldoc.getElementsByTagName("kml")[0]
document = kml.getElementsByTagName("Document")[0]
folder = document.getElementsByTagName("Folder")[0]
placemarks = folder.getElementsByTagName("Placemark")

longitudeList = []
latitudeList = []
buildingTypeList = []

for placemark in placemarks:
    # Extract type of the house
    #green: singleHouse, red: garage, yellow: commercial_building, pink: collective_house, blue:light_building, cyan:school
    color = placemark.getElementsByTagName("styleUrl")[0].firstChild.data
    if(color == "#green"):
        buildingType = "single_house"
    elif(color == "#red"):
        buildingType = "garage"
    elif(color == "#yellow"):
        buildingType = "commercial_building"
    elif(color == "#pink"):
        buildingType = "collective_house"
    elif(color == "#blue"):
        buildingType = "light_building"
    elif(color == "#cyan"):
        buildingType = "school"
    
    # Extract coordinates
    coords = placemark.getElementsByTagName("coordinates")[0].firstChild.data
    lst1 = coords.split(" ")
    lst2 = lst1[0].split(",")
    longitude = float(lst2[0])
    latitude = float(lst2[1])
    longitudeList.append(longitude)
    latitudeList.append(latitude)
    buildingTypeList.append(buildingType)
    
   # print("longitude: ", longitude,"latitude: ", latitude, "building_type: ", buildingType)
    
df = pd.DataFrame(list(zip(longitudeList, latitudeList, buildingTypeList)), 
               columns =['longitude', 'latitude', 'buildingType']) 
```

```python
#Previous approach couldn't read the whole data set.

from bs4 import BeautifulSoup

fileName = 'Zone_3.kml'

with open(fileName) as data:
    kml_soup = BeautifulSoup(data, 'lxml-xml') # Parse as XML

placemarks = kml_soup.find_all('Placemark')

longitudeList = []
latitudeList = []
buildingTypeList = []
```

```python
for placemark in placemarks:
    # Extract building type
    color = placemark.styleUrl.string
    if(color == "#green"):
        buildingType = "single_house"
    elif(color == "#red"):
        buildingType = "garage"
    elif(color == "#yellow"):
        buildingType = "commercial_building"
    elif(color == "#pink"):
        buildingType = "collective_house"
    elif(color == "#blue"):
        buildingType = "light_building"
    elif(color == "#cyan"):
        buildingType = "school"
    
    # Extract coordinates
    coords = placemark.Polygon.outerBoundaryIs.LinearRing.coordinates.string
    lst1 = coords.split(" ")
    lst2 = lst1[0].split(",")
    longitude = float(lst2[0])
    latitude = float(lst2[1])
    longitudeList.append(longitude)
    latitudeList.append(latitude)
    buildingTypeList.append(buildingType)
    
    
df = pd.DataFrame(list(zip(longitudeList, latitudeList, buildingTypeList)), 
               columns =['longitude', 'latitude', 'buildingType'])

df.to_csv('Zone_3.csv')
```
